## Exploratory Data Analysis (EDA) with a Focus on Static Visualization:

```python
# Visualization 1 : Analytical purpose is to provide a clear and concise overview of property distribution across the top 20 towns, aiding in data exploration and pattern recognition.
#The bar chart visually represents the distribution of properties across the top 20 towns.
#It allows for a quick comparison of the property count in each town, identifying towns with higher and lower property counts.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame

# Get the top 20 towns
top_20_towns = df['Town'].value_counts().nlargest(20).index

# Filter the DataFrame for the top 20 towns
df_top_20 = df[df['Town'].isin(top_20_towns)]

# Bar chart for Top 20 Town distribution
plt.figure(figsize=(14, 8))
sns.countplot(x='Town', data=df_top_20, order=top_20_towns)
plt.title('Distribution of Top 20 Towns')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 2: Time Series Analysis: Assessed Value over the years
plt.figure(figsize=(12, 6))
sns.lineplot(x='List Year', y='Assessed Value', data=df)
plt.title('Trend of Assessed Value Over the Years')
plt.xlabel('List Year')
plt.ylabel('Assessed Value')
plt.show()
```

```python
# Visualization 3 : Scatter Plot with Size Variation based on Assessed Value
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


plt.figure(figsize=(14, 8))
sns.scatterplot(x='Assessed Value', y='Sale Amount', hue='Type1', size='Sales Ratio', sizes=(20, 200), palette='viridis', data=df)
plt.title('Scatter Plot of Assessed Value vs Sale Amount (Colored by Type1, Sized by Sales Ratio)')
plt.xlabel('Assessed Value')
plt.ylabel('Sale Amount')
plt.legend(title='Type1', bbox_to_anchor=(1.05, 1), loc='upper left')  # Move the legend outside the plot
plt.show()
```

```python
#Visualization 4 : Assessed value Across Top 20 Towns

import matplotlib.pyplot as plt
import seaborn as sns

top_20_towns = df.groupby('Town')['Assessed Value'].mean().nlargest(20).index

# Filter the DataFrame for the top 20 towns
df_top_20 = df[df['Town'].isin(top_20_towns)]

# Violin plot for Assessed Value across Top 20 Towns
plt.figure(figsize=(14, 8))
sns.violinplot(x='Town', y='Assessed Value', data=df_top_20, order=top_20_towns)
plt.title('Violin Plot of Assessed Value Across Top 20 Towns')
plt.xlabel('Town')
plt.ylabel('Assessed Value')
plt.xticks(rotation=45)
plt.show()
```

## Advanced Analysis:

```python
# Time series analysis for 'Date Recorded'

df['Date Recorded'] = pd.to_datetime(df['Date Recorded'])
df.set_index('Date Recorded', inplace=True)
monthly_sales = df.resample('M').sum()
plt.figure(figsize=(14, 8))
monthly_sales['Sale Amount'].plot()
plt.title('Monthly Sales Amount Over Time')
plt.xlabel('Date')
plt.ylabel('Sale Amount')
plt.show()
```

```python
#Decomposition: Decomposing the time series into components like trend, seasonality, and residual.

import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Resample the data to ensure there are no missing dates
df_resampled = df.resample('M').sum()


# Perform seasonal decomposition
result = seasonal_decompose(df_resampled['Sale Amount'], model='additive')


# Plot the components
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(result.observed, label='Observed')
plt.legend()


plt.subplot(4, 1, 2)
plt.plot(result.trend, label='Trend')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(result.seasonal, label='Seasonal')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(result.resid, label='Residual')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Forecasting Models: Building sophisticated forecasting models using methods like ARIMA, SARIMA, or machine learning algorithms.

import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt


# Resample the data to ensure there are no missing dates
df_resampled = df.resample('M').sum()

# Fit ARIMA model
order = (5, 1, 2)  # Example order, you may need to tune these parameters
model = ARIMA(df_resampled['Sale Amount'], order=order)
results = model.fit()

# Forecast future values
forecast_steps = 12  # Number of steps to forecast
forecast = results.get_forecast(steps=forecast_steps)

# Plot the original data and forecast
plt.figure(figsize=(14, 8))
plt.plot(df_resampled['Sale Amount'], label='Original Data')
plt.plot(forecast.predicted_mean, label='Forecasted Data', color='red')
plt.title('ARIMA Forecasting')
plt.xlabel('Date')
plt.ylabel('Sale Amount')
plt.legend()
plt.show()
```

```python
#Anomaly Detection: Identifying anomalies or outliers in the time series data.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Resample the data to ensure there are no missing dates
df_resampled = df.resample('M').sum()

# Calculate z-scores
mean = df_resampled['Sale Amount'].mean()
std_dev = df_resampled['Sale Amount'].std()
z_scores = (df_resampled['Sale Amount'] - mean) / std_dev

# Define a threshold for anomaly detection (e.g., z-score > 3)
threshold = 3

# Identify anomalies
anomalies = df_resampled[z_scores.abs() > threshold]

# Plot the original data and highlight anomalies
plt.figure(figsize=(14, 8))
plt.plot(df_resampled['Sale Amount'], label='Original Data')
plt.scatter(anomalies.index, anomalies['Sale Amount'], color='red', label='Anomalies')
plt.title('Anomaly Detection in Time Series Data')
plt.xlabel('Date')
plt.ylabel('Sale Amount')
plt.legend()
plt.show()
```

```python
#Advanced Seasonal Decomposition: Applying more advanced decomposition methods like STL (Seasonal-Trend decomposition using LOESS)

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL


# Resample the data to ensure there are no missing dates
df_resampled = df.resample('M').sum()

# Perform STL decomposition
stl = STL(df_resampled['Sale Amount'], period=12)
result = stl.fit()

# Plot the components
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(14, 10))

ax1.plot(result.trend, label='Trend')
ax1.legend()

ax2.plot(result.seasonal, label='Seasonal')
ax2.legend()

ax3.plot(result.resid, label='Residual')
ax3.legend()

plt.tight_layout()
plt.show()
```
